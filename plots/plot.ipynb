```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# Data is Epoch/Accuracy/Loss/Val_accuracy/Val_loss
train_data = np.genfromtxt(os.path.join(os.pardir,'logs', 'vgg_baseline_attention_model_256.csv'), delimiter=',')
train_data = train_data[1:]
```

```python
# Data is Epoch/Accuracy/Loss/Val_accuracy/Val_loss
res = {}
for i in [17, 42, 90, 256]:
        train_data = np.genfromtxt(os.path.join(os.pardir,'logs', f'vgg_baseline_attention_model_{i}.csv'), delimiter=',')
        res[i] = train_data[1:]
```

```python
fig, ax = plt.subplots(figsize=(8,6))
for i in [17, 42, 90, 256]:
    x = res[i][:, 0]
    ax.plot(x, res[i][:, 1], '-', label=f'Classification - Training Accuracy @{i}')
ax.legend()
ax.set_title('Training Metrics (Accuracy)')
ax.set_xlabel('Epochs')
ax.set_ylabel('Accuracy')
ax.grid()
#plt.savefig('acc_per_epoch_base_30.png')
```

```python
fig, ax = plt.subplots(figsize=(8,6))
for i in [17, 42, 90, 256]:
    x = res[i][:, 0]
    ax.plot(x, res[i][:, 3], '--', label=f'Classification - Validation Accuracy @{i}')
ax.legend()
ax.set_title('Training Metrics (Accuracy)')
ax.set_xlabel('Epochs')
ax.set_ylabel('Accuracy')
ax.grid()
#plt.savefig('acc_per_epoch_base_30.png')
```

```python
fig, ax = plt.subplots(figsize=(8,6))
x = train_data[:, 0]
ax.plot(x, train_data[:, 1], 'r-', label='Classification - Training Accuracy')
ax.plot(x, train_data[:, 3], 'r--', label='Classification - Validation Accuracy')
ax.legend()
ax.set_title('Training Metrics (Accuracy) for dim 30')
ax.set_xlabel('Epochs')
ax.set_ylabel('Accuracy')
ax.grid()
#plt.savefig('acc_per_epoch_base_30.png')
```

```python
fig, ax = plt.subplots(figsize=(8,6))
x = train_data[:, 0]
ax.plot(x, train_data[:, 2], 'b-', label='Classification - Training Loss')
ax.plot(x, train_data[:, 4], 'b--', label='Classification - Validation Loss')
ax.legend()
ax.set_title('Training Metrics (Loss) for dim 42')
ax.set_xlabel('Epochs')
ax.set_ylabel('Loss')
ax.grid()
plt.savefig('loss_per_epoch_base_42.png')
```
