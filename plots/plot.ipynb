```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# Data is Epoch/Accuracy/Loss/Val_accuracy/Val_loss
train_data = np.genfromtxt(os.path.join(os.pardir,'logs', 'vgg_training_att.csv'), delimiter=',')
train_data = train_data[1:]
```

```python
fig, ax = plt.subplots(figsize=(8,6))
x = train_data[:, 0]
ax.plot(x, train_data[:, 1], 'r-', label='Classification - Training Accuracy')
ax.plot(x, train_data[:, 3], 'r--', label='Classification - Validation Accuracy')
ax.legend()
ax.set_xlabel('Epochs')
ax.set_ylabel('Accuracy')
ax.grid()
plt.savefig('acc_per_epoch.png')
```

```python
fig, ax = plt.subplots(figsize=(8,6))
x = train_data[:, 0]
ax.plot(x, train_data[:, 2], 'b-', label='Classification - Training Loss')
ax.plot(x, train_data[:, 4], 'b--', label='Classification - Validation Loss')
ax.legend()
ax.set_xlabel('Epochs')
ax.set_ylabel('Loss')
ax.grid()
plt.savefig('loss_per_epoch.png')
```
